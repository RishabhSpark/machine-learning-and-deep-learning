#### (a)

```python
df_2 = pd.read_csv('diabetes.csv')
df_2.head()
```

```python
df_2.agg({'Glucose': ['mean', 'std']})
```

#### (b)

```python
df_2.groupby('Outcome')['BMI'].median()
```

```python
df_2['BMI'] = df_2['BMI'].fillna(df_2.groupby('Outcome')['BMI'].transform('median'))
```

#### (c)

```python
Age_Category = []
for row in df_2['Age']:
    if row >=50 :    Age_Category.append('Old')
    elif row <50 & row>21:   Age_Category.append('Middle_Young')
    elif row == 21:  Age_Category.append('Young')
    else : Age_Category.append("Below 21")
    
df_2['Age_Category'] = Age_Category
df_2
```

#### (d)

```python
import numpy as np
cols = ["Glucose","BloodPressure","SkinThickness","Insulin"]
df_2[cols] = df_2[cols].replace({'0':np.nan, 0:np.nan})
```

```python
df_2.isnull().sum()
```

```python
df_2[cols]=df_2[cols].fillna(df_2.mean().iloc[0])
df_2
```
